**Importing the Libraries**

```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
train_data.info()
```

```python
train_data
```

```python
columns_to_drop = ['PassengerId', 'Name', 'Ticket', 'Survived', 'Cabin']
numerical_features = train_data.drop(columns_to_drop, axis = 1).select_dtypes(include = ['number']).columns.tolist()
categorical_features = train_data.drop(columns_to_drop, axis = 1).select_dtypes(exclude = ['number']).columns.tolist()
```

```python
numerical_features
```

```python
categorical_features
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(train_data[numerical_features])
train_data[numerical_features] = imputer.transform(train_data[numerical_features])
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(train_data[categorical_features])
train_data[categorical_features] = imputer.transform(train_data[categorical_features])
```

```python
y = train_data.iloc[:, 1].values
```

```python
y
```

```python
train_data = train_data.drop('Survived', axis=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), categorical_features)], remainder = 'passthrough')
train_data = ct.fit_transform(train_data)
```

```python
# Convert the transformed dataset back to a DataFrame
train_data = pd.DataFrame(train_data)

# Print the first few rows of the transformed DataFrame
print(train_data.head())
```

```python
train_data.info()
```

```python
X = train_data.iloc[:, [0] + list(range(1, 7)) + list(range(8, 11)) + [12]].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
columns_to_scale = [5, 6, 7, 10]
x_train[:,columns_to_scale] = sc.fit_transform(x_train[:,columns_to_scale])
```

```python
x_test[:,columns_to_scale] = sc.transform(x_test[:,columns_to_scale])
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(random_state = 1)
lr_classifier.fit(x_train, y_train)
```

```python
lr_y_pred = lr_classifier.predict(x_test)
```

```python
print(np.concatenate((lr_y_pred.reshape(len(lr_y_pred), 1),
                     y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, lr_y_pred)
print(cm)
accuracy_score(y_test, lr_y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2)
knn_classifier.fit(x_train, y_train)
```

```python
knn_y_pred = knn_classifier.predict(x_test)
print(np.concatenate((knn_y_pred.reshape(len(knn_y_pred), 1),
                     y_test.reshape(len(y_test), 1)), 1))
```

```python
cm = confusion_matrix(y_test, knn_y_pred)
print(cm)
accuracy_score(y_test, knn_y_pred)
```

```python
from sklearn.svm import SVC
svm_classifier = SVC(kernel = 'linear', random_state = 1)
svm_classifier.fit(x_train, y_train)
```

```python
svm_y_pred = svm_classifier.predict(x_test)
print(np.concatenate((svm_y_pred.reshape(len(svm_y_pred), 1),
                     y_test.reshape(len(y_test), 1)), 1))
```

```python
cm = confusion_matrix(y_test, svm_y_pred)
print(cm)
accuracy_score(y_test, svm_y_pred)
```

```python
from sklearn.svm import SVC
ksvm_classifier = SVC(kernel = 'rbf', random_state = 1)
ksvm_classifier.fit(x_train, y_train)
```

```python
ksvm_y_pred = ksvm_classifier.predict(x_test)
print(np.concatenate((ksvm_y_pred.reshape(len(ksvm_y_pred), 1),
                     y_test.reshape(len(y_test), 1)), 1))
```

```python
cm = confusion_matrix(y_test, ksvm_y_pred)
print(cm)
accuracy_score(y_test, ksvm_y_pred)
```
